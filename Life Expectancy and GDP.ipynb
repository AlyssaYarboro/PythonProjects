## Life Expectancy and GDP in Different Countries

## Modifying the Data Frame

##### Let's modify the DataFrame columns so they are a little easier to work with. All the columns are either one word or an abbreviated. Let's change the column 'Life expectancy at birth (years)' to a shorthand so it's easier to work with in the next steps.

```python
df.rename(columns = {"Life expectancy at birth (years)": "LEABY"}, inplace = True)
```

```python
print(df.head())
```

## Creating A Violin Plot

```python
fig = plt.subplots(figsize=(15, 20))
sns.violinplot(x = 'Country', y = 'LEABY', data = df)
plt.show()
```

## Creating Bar Plots

```python
ax = sns.barplot(x = 'Country', y = 'LEABY', data = df)
plt.xticks(rotation=90)
plt.show()
```
